```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edge('Fabrika', 'Depo A', weight=10)
G.add_edge('Fabrika', 'Depo B', weight=20)
G.add_edge('Depo A', 'Şantiye', weight=15)
G.add_edge('Depo B', 'Şantiye', weight=5)
G.add_edge('Depo A', 'Depo B', weight=8)

path = nx.dijkstra_path(G, 'Fabrika', 'Şantiye', weight='weight')
length = nx.dijkstra_path_length(G, 'Fabrika', 'Şantiye', weight='weight')

print(f"En Verimli Rota: {path}")
print(f"Toplam Mesafe: {length}")

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='orange', node_size=2500, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Lojistik Rota Optimizasyonu")
plt.show()
```
